```python
# Import necessary libraries for data analysis and visualization
import pandas as pd  # For data manipulation
import numpy as np  # For numerical operations
import seaborn as sns  # For creating statistical visualizations

import matplotlib.pyplot as plt  # For plotting graphs
import matplotlib  # For controlling various plotting parameters

# Set the default style for visualizations
plt.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (12, 8)  # Set the figure size for all plots

# Read in the dataset
df = pd.read_csv('movies.csv')  # Load the movie dataset

# Disable chained assignment warnings (to avoid unnecessary messages)
pd.options.mode.chained_assignment = None
```

```python
# Now let's take a look at the data
df
```

```python
# Check for any missing data in the dataset
# We loop through each column to calculate the percentage of missing values

for column in df.columns:
    missing_percentage = df[column].isnull().mean() * 100  # Calculate percentage of missing values
    print(f'{column}: {missing_percentage:.2f}% missing')  # Print result with two decimal places
```

```python
# Create a scatter plot to visualize the relationship between 'gross' and 'budget'
# This helps us understand how movie budgets relate to their box office performance

sns.regplot(x="gross", y="budget", data=df)  # Plot with a regression line for better insights
```

```python
# Create a scatter plot to analyze the relationship between 'score' (ratings) and 'gross' (box office earnings)
# This helps us see if higher-rated movies tend to make more money

sns.regplot(x="score", y="gross", data=df)  # Adds a regression line for clarity
```

```python
# Correlation Matrix between all numeric columns
df.corr(method ='pearson', numeric_only=True) # add numeric_only argument
```

```python
# Calculate the correlation matrix, but only include numeric columns to avoid the error.
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value
df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
# Create a correlation matrix for the movie dataset
# Factorize the features to numeric values and compute the Pearson correlation

correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

# Visualize the correlations between movie features using a heatmap

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")  # Display correlation values on the heatmap
plt.title("Correlation Matrix for Movie Features")  # Set the title of the plot
plt.xlabel("Movie Features")  # Label for the x-axis
plt.ylabel("Movie Features")  # Label for the y-axis
plt.show()  # Display the plot
```

```python
# Generate a correlation matrix by factorizing the movie features into numerical values
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr()

# Unstack the matrix to see all the feature pairs and their correlation values
correlation_pairs = correlation_matrix.unstack()

# Print out all feature pairs along with their correlation values
print(correlation_pairs)
```

```python
# Sort the feature correlation pairs in ascending order
sorted_correlation_pairs = corr_pairs.sort_values(kind="quicksort")

# Display the sorted correlation pairs
print(sorted_correlation_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]
print(strong_pairs)
```

```python
# Analyze the top 15 companies by total gross revenue
top_15_companies = df.groupby('company')[["gross"]].sum()  # Sum up gross revenue for each company
top_15_companies_sorted = top_15_companies.sort_values('gross', ascending=False).head(15)  # Sort and get the top 15
top_15_companies_sorted['gross'] = top_15_companies_sorted['gross'].astype(int)  # Convert gross revenue to integer
top_15_companies_sorted
```

```python
# Analyze the top 15 companies by gross revenue over the years
top_15_company_years = df.groupby(['company', 'year'])[['gross']].sum()  # Summing up gross revenue per company by year
top_15_sorted = top_15_company_years.sort_values(['gross', 'company', 'year'], ascending=False).head(15)  # Sort by revenue, company, and year
top_15_sorted['gross'] = top_15_sorted['gross'].astype(int)  # Convert gross revenue to integers
top_15_sorted
```

```python
# Group data by 'company' and sum their gross earnings
top_companies = df.groupby('company')['gross'].sum()

# Sort the companies by gross earnings in descending order and select the top 15
top_companies_sorted = top_companies.sort_values(ascending=False).head(15)

# Convert the 'gross' column to integer for clarity
top_companies_sorted = top_companies_sorted.astype('int64')

# Display the result
top_companies_sorted
```
